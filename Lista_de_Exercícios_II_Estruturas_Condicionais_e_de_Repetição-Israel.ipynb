```python


salario_atual_str = input("Digite o salário atual do funcionário ")
salario_atual = float(salario_atual_str)

percentual_aumento = 0
novo_salario = 0

if salario_atual <= 2000.00:
    percentual_aumento = 0.10
elif salario_atual <= 5000.00:
    percentual_aumento = 0.07
else:
    percentual_aumento = 0.05


aumento = salario_atual * percentual_aumento
novo_salario = salario_atual + aumento


print(f"\nSalário atual: R$ {salario_atual:.2f}")
print(f"Percentual de aumento aplicado: {percentual_aumento * 100:.0f}%")
print(f"Valor do aumento: R$ {aumento:.2f}")
print(f"Novo salário: R$ {novo_salario:.2f}")
```

```python
soma = 0
quantidade = 0

idade = int(input("Digite uma idade (valor negativo para parar): "))

while idade >= 0:
    soma += idade
    quantidade += 1
    idade = int(input("Digite outra idade (valor negativo para parar): "))

if quantidade > 0:
    media = soma / quantidade
    print("Média das idades:", media)
else:
    print("Nenhuma idade válida foi digitada.")
```

```python
positivos = 0
negativos = 0
zeros = 0

for i in range(10):
    num = float(input(f"Digite o {i+1}º número: "))

    if num > 0:
        positivos += 1
    elif num < 0:
        negativos += 1
    else:
        zeros += 1

print("Positivos:", positivos)
print("Negativos:", negativos)
print("Zeros:", zeros)
```
