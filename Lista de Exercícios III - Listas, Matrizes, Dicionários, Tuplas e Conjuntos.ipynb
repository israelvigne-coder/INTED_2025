```python
#13. Solicite uma matriz 3x3 preenchida pelo usuário. Verifique se ela é uma matriz identidade.

def verificar_matriz_identidade(matriz):
    n = len(matriz)
    for i in range(n):
        for j in range(n):
            if i == j:
                if matriz[i][j] != 1:
                    return False
            else:
                if matriz[i][j] != 0:
                    return False
    return True


matriz = []
print("Digite os valores para a matriz 3x3:")
for i in range(3):
    linha = []
    for j in range(3):
        while True:
            try:
                valor = int(input(f"Digite o valor para a posição [{i+1}][{j+1}]: "))
                linha.append(valor)
                break
            except ValueError:
                print("Entrada inválida. Por favor, digite um número inteiro.")
    matriz.append(linha)

print("\nMatriz digitada:")
for linha in matriz:
    print(linha)


if verificar_matriz_identidade(matriz):
    print("\nA matriz é uma matriz identidade.")
else:
    print("\nA matriz NÃO é uma matriz identidade.")
```

```python
#14. Faça um programa que leia uma matriz 3x3 e verifique se ela é simétrica (igual à sua transposta).


def ler_matriz():
    matriz = []
    print("Digite os valores para a matriz 3x3:")
    for i in range(3):
        linha = []
        for j in range(3):
            while True:
                try:
                    valor = int(input(f"Digite o valor para a posição [{i+1}][{j+1}]: "))
                    linha.append(valor)
                    break
                except ValueError:
                    print("Entrada inválida. Por favor, digite um número inteiro.")
        matriz.append(linha)
    return matriz


def transpor_matriz(matriz):
    n = len(matriz)
    transposta = [[0 for _ in range(n)] for _ in range(n)]
    for i in range(n):
        for j in range(n):
            transposta[j][i] = matriz[i][j]
    return transposta


def verificar_simetria(matriz):
    transposta = transpor_matriz(matriz)
    return matriz == transposta
matriz_usuario = ler_matriz()

print("\nMatriz digitada:")
for linha in matriz_usuario:
    print(linha)

if verificar_simetria(matriz_usuario):
    print("\nA matriz é simétrica.")
else:
    print("\nA matriz NÃO é simétrica.")
```

```python
#17. Dada uma matriz 3x3 com valores reais, calcule o determinante (sem usar bibliotecas externas).


def ler_matriz():
    matriz = []
    print("Digite os valores para a matriz 3x3:")
    for i in range(3):
        linha = []
        for j in range(3):
            while True:
                try:
                    valor = float(input(f"Digite o valor para a posição [{i+1}][{j+1}]: "))
                    linha.append(valor)
                    break
                except ValueError:
                    print("Entrada inválida. Por favor, digite um número real.")
        matriz.append(linha)
    return matriz


def calcular_determinante_3x3(matriz):

    a, b, c = matriz[0][0], matriz[0][1], matriz[0][2]
    d, e, f = matriz[1][0], matriz[1][1], matriz[1][2]
    g, h, i = matriz[2][0], matriz[2][1], matriz[2][2]

    determinante = a * (e * i - f * h) - b * (d * i - f * g) + c * (d * h - e * g)
    return determinante


matriz_usuario = ler_matriz()

print("\nMatriz digitada:")
for linha in matriz_usuario:
    print(linha)

determinante = calcular_determinante_3x3(matriz_usuario)
print(f"\nO determinante da matriz é: {determinante}")
```

```python
#18. Faça um programa que leia duas matrizes 3x3 e calcule a matriz soma e a matriz diferença.

def ler_matriz(nome_matriz):

    matriz = []
    print(f"\nDigite os valores para a {nome_matriz} (matriz 3x3):")
    for i in range(3):
        linha = []
        for j in range(3):
            while True:
                try:
                    valor = float(input(f"Digite o valor para a posição [{i+1}][{j+1}]: "))
                    linha.append(valor)
                    break
                except ValueError:
                    print("Entrada inválida. Por favor, digite um número real.")
        matriz.append(linha)
    return matriz


def calcular_soma_matrizes(matriz1, matriz2):
    matriz_soma = [[0 for _ in range(3)] for _ in range(3)]
    for i in range(3):
        for j in range(3):
            matriz_soma[i][j] = matriz1[i][j] + matriz2[i][j]
    return matriz_soma


def calcular_diferenca_matrizes(matriz1, matriz2):
    matriz_diferenca = [[0 for _ in range(3)] for _ in range(3)]
    for i in range(3):
        for j in range(3):
            matriz_diferenca[i][j] = matriz1[i][j] - matriz2[i][j]
    return matriz_diferenca


matriz_a = ler_matriz("Matriz A")
matriz_b = ler_matriz("Matriz B")

matriz_soma = calcular_soma_matrizes(matriz_a, matriz_b)
matriz_diferenca = calcular_diferenca_matrizes(matriz_a, matriz_b)

print("\n----- Matrizes Digitadas -----")
print("Matriz A:")
for linha in matriz_a:
    print(linha)
print("\nMatriz B:")
for linha in matriz_b:
    print(linha)

print("\n----- Resultados -----")
print("Matriz Soma (A + B):")
for linha in matriz_soma:
    print(linha)

print("\nMatriz Diferença (A - B):")
for linha in matriz_diferenca:
    print(linha)
```

```python
#24. Dado um conjunto de números inteiros, crie um novo conjunto com apenas os quadrados perfeitos encontrados.

import math

def is_perfect_square(num):
    if num < 0:
        return False
    if num == 0:
        return True
    sqrt_num = int(math.sqrt(num))
    return sqrt_num * sqrt_num == num


def ler_conjunto_inteiros(nome_conjunto):
    while True:
        entrada = input(f"Digite os elementos do {nome_conjunto} separados por vírgula (ex: 1,4,9,10): ")
        try:
            elementos = {int(x.strip()) for x in entrada.split(',')}
            return elementos
        except ValueError:
            print("Entrada inválida. Por favor, digite apenas números inteiros separados por vírgula.")

conjunto_original = ler_conjunto_inteiros("conjunto de números inteiros")

conjunto_quadrados_perfeitos = set()

for numero in conjunto_original:
    if is_perfect_square(numero):
        conjunto_quadrados_perfeitos.add(numero)

print(f"\nConjunto original: {conjunto_original}")
print(f"Conjunto de quadrados perfeitos: {conjunto_quadrados_perfeitos}")
```

```python
#25. Crie um dicionário que armazene o nome e o salário de 5 funcionários. Calcule o reajuste de 10% para salários inferiores a R$ 2.000 e atualize o dicionário.


funcionarios = {}
print("Digite o nome e o salário de 5 funcionários:")
for i in range(5):
    while True:
        nome = input(f"Digite o nome do {i+1}º funcionário: ")
        if nome.strip() == '':
            print("Nome não pode ser vazio. Por favor, digite novamente.")
            continue
        try:
            salario = float(input(f"Digite o salário de {nome}: R$ "))
            if salario < 0:
                print("Salário não pode ser negativo. Por favor, digite novamente.")
                continue
            funcionarios[nome] = salario
            break
        except ValueError:
            print("Entrada inválida para o salário. Por favor, digite um número.")

print("\nDicionário de funcionários original:")
for nome, salario in funcionarios.items():
    print(f"Nome: {nome}, Salário: R$ {salario:.2f}")


for nome, salario in funcionarios.items():
    if salario < 2000:
        funcionarios[nome] = salario * 1.10

print("\nDicionário de funcionários atualizado após reajuste:")
for nome, salario in funcionarios.items():
    print(f"Nome: {nome}, Salário: R$ {salario:.2f}")
```

```python
#27. Crie um dicionário que armazene palavras e seus sinônimos. Depois, permita que o usuário digite uma palavra e veja seus sinônimos (se existirem).


sinonimos = {
    "feliz": ["alegre", "contente", "satisfeito"],
    "triste": ["infeliz", "melancólico", "abatido"],
    "grande": ["enorme", "imenso", "vasto"],
    "pequeno": ["minúsculo", "diminuto", "míni"],
    "rápido": ["veloz", "ligeiro", "acelerado"]
}

print("Dicionário de sinônimos carregado.")

while True:
    palavra = input("\nDigite uma palavra para buscar sinônimos (ou 'sair' para encerrar): ").lower()

    if palavra == 'sair':
        print("Programa encerrado.")
        break

    if palavra in sinonimos:
        print(f"Sinônimos para '{palavra}': {', '.join(sinonimos[palavra])}")
    else:
        print(f"Nenhum sinônimo encontrado para '{palavra}'.")
```

```python
#28. Solicite ao usuário 10 palavras e armazene em um dicionário onde a chave é a primeira letra e o valor é uma lista de palavras que começam com essa letra.


dicionario_palavras = {}

print("Digite 10 palavras:")
for i in range(10):
    palavra = input(f"Digite a {i+1}ª palavra: ").strip()
    if palavra:
        primeira_letra = palavra[0].lower()


        if primeira_letra in dicionario_palavras:
            dicionario_palavras[primeira_letra].append(palavra)
        else:

            dicionario_palavras[primeira_letra] = [palavra]

print("\nDicionário de palavras por primeira letra:")
for letra, palavras_lista in dicionario_palavras.items():
    print(f"'{letra}': {palavras_lista}")
```

```python
#29. Crie um dicionário onde as chaves são nomes de produtos e os valores são listas com preços praticados em três lojas diferentes. Encontre o produto com maior variação de preço.


produtos = {
    "Camiseta": [25.00, 29.90, 22.50],
    "Calça Jeans": [89.90, 95.00, 79.99],
    "Tênis": [120.00, 135.00, 115.00],
    "Relógio": [50.00, 65.00, 55.00],
    "Mochila": [70.00, 72.00, 68.00]
}

maior_variacao = -1
produto_maior_variacao = ""

for produto, precos in produtos.items():
    if precos:
        variacao_atual = max(precos) - min(precos)
        if variacao_atual > maior_variacao:
            maior_variacao = variacao_atual
            produto_maior_variacao = produto

print("Dicionário de produtos e preços:")
for produto, precos in produtos.items():
    print(f"  {produto}: {precos}")

if produto_maior_variacao:
    print(f"\nO produto com a maior variação de preço é '{produto_maior_variacao}' com uma variação de R$ {maior_variacao:.2f}.")
else:
    print("Nenhum produto encontrado no dicionário.")
```

```python
#30. Implemente um programa que leia uma frase e armazene em um dicionário o número de ocorrências de cada palavra (ignorando letras maiúsculas/minúsculas).

frase = input("Digite uma frase: ")


frase_minuscula = frase.lower()


import string
tabela_traducao = str.maketrans('', '', string.punctuation)
frase_sem_pontuacao = frase_minuscula.translate(tabela_traducao)

palavras = frase_sem_pontuacao.split()


ocorrencias_palavras = {}

for palavra in palavras:
    if palavra:
        ocorrencias_palavras[palavra] = ocorrencias_palavras.get(palavra, 0) + 1

print("\nNúmero de ocorrências de cada palavra:")
for palavra, contagem in ocorrencias_palavras.items():
    print(f"'{palavra}': {contagem}")
```
